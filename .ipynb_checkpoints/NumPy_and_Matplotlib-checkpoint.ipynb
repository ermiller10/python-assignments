```python
import numpy as np
```

```python
x = np.arange(-2.0,2.02,0.02)
```

```python
y = np.linspace(-2,2,200)
```

```python
from matplotlib import pyplot as plt
```

```python

xx, yy = np.meshgrid(x, y)
xx.shape, yy.shape

plt.pcolormesh(yy)
```

```python
plt.pcolormesh(xx, yy, xx)
```

```python
r = np.sqrt(xx**2 + yy**2)
```

```python

phi = np.arctan2(yy, xx)
```

```python
fig, ax1 = plt.subplots()
img = ax1.pcolormesh(xx, yy, r)
fig.colorbar (img, ax=ax1)
```

```python
f = cos**2(4*r) + sin**2(4*phi)
```

```python
plt.plot(x, y)
plt.show(f)
```

```python
#1.6 plot mean with respect to one aspect
r_mean = r.mean(axis=0)
```
